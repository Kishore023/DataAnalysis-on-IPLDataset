## Similarly we can create a function to do all this tasks for two particular teams

```python
def two_team_statistics(team1, team2):
    data = df[((df['team1']== team1) | (df['team1']== team2 )) & ((df['team2']== team2) | (df['team2']== team1))]
    fig, ax =plt.subplots(2,2, figsize =(15,10))
    sns.countplot(data['winner'], ax =ax[0,0])
    sns.countplot(x='toss_winner', hue= 'winner' , ax= ax[0,1], data = data)
    sns.countplot('winner', hue = 'city',ax = ax[1,0], data = data)
    ax[1,1].pie(data['winner'].value_counts(), labels = [team1, team2], autopct = '%1.1f%%', shadow = True)
    ax[1,1].legend(loc = 'center left')
    plt.show()
two_team_statistics('Royal Challengers Bangalore', 'Mumbai Indians')
```

```python
two_team_statistics('Royal Challengers Bangalore', 'Delhi Daredevils')
```
